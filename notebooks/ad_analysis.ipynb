```python
# Load AD results

data_dir = '../data/'
table = load_ad_tables_by_patterns([os.path.join(data_dir, 'm31_iso.csv'),
                                    os.path.join(data_dir, 'm31_gmm.csv'),
                                    os.path.join(data_dir, 'm31_lof.csv'),
                                    os.path.join(data_dir, 'm31_svm.csv'),])
# Sort the table
table = table.loc[table.isna().sum(axis=1).sort_values(kind='mergesort').index].reset_index(drop=True)

# Load experts' evaluations
expert_table = pd.read_csv(os.path.join(data_dir, 'maria.csv'), usecols=np.arange(4), names=['oid', 'name', 'comment', 'value'])
```

# AD results comparison

```python
# subtable = extract_ad_subtable(table, [c for c in table.columns if c.endswith('_ztf')])
subtable = pd.merge(table, expert_table, on='oid', how='left')
display(subtable)
```

```python
# Download coordinates of objects

import requests
from tqdm import tqdm

for column in ['ra', 'dec']:
    if column not in subtable:
        subtable[column] = None

for i in tqdm(subtable.index):
    oid = str(subtable['SNe'][i])
    resp = requests.get(f'http://db.ztf.snad.space/api/v1/oid/full/json?oid={oid}')
    coord = resp.json()[oid]['meta']['coord']
    subtable.loc[i, 'ra'] = coord['ra']
    subtable.loc[i, 'dec'] = coord['dec']
```

# Celestial plot

```python
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(111, projection='lambert')
ax.grid(True)
sct = ax.scatter(subtable['ra'] * np.pi / 180,
                 subtable['dec'] * np.pi / 180,
                 marker='*',
                 c=subtable.index, cmap='viridis')

fig.colorbar(sct, ax=ax)
pass
```
